```python
def unemply_func(row): 
    if row == '3-4.9%':
        return 10
    elif row == '15-19.9%':
        return 5.01
    elif row == '5-6.9%':
        return 4.87
    elif row == '11-14.9%':
        return 1.26
    elif row == '9-10.9%':
        return 0.86
    elif row == '7-8.9%': 
        return 0.44
    else: 
        return 0
```

```python
def edu_college_func(row): 
    if row == '5-10%':
        return 7.58
    elif row == '10-15%':
        return 3.06
    elif row == '15-20%':
        return 1.86
    elif row == '50-55%':
        return 1.86
    elif row == '40-45%':
        return 1.36
    elif row == '45-50%': 
        return 0.79
    elif row == '30-35%':
        return 0.70
    elif row == '60-65%':
        return 0.62
    else: 
        return 0
```

```python
def edu_highschool_func(row): 
    if row == '15-20%':
        return 2.35
    elif row == '30-35%':
        return 0.92
    elif row == '10-15%':
        return 0.41
    elif row == '45-50%':
        return 0.40
    else: 
        return 0
```

```python
def labor_func(row): 
    if row == '65-67.9%':
        return 6.28
    elif row == '62-64.9%':
        return 5.98
    elif row == '53-55.9%':
        return 3
    elif row == '50-52.9%%':
        return 1.58
    elif row == '59-61.9%':
        return 0.56
    elif row == '56-58.9%':
        return 0.48
    elif row == '40-49.9%':
        return 0.46
    else: 
        return 0
```

```python
def income_func(row): 
    if row == '35k_to_50k':
        return 4.02
    elif row == '50k_to_75k':
        return 3.15
    elif row == '75k_to_100k':
        return 2.19
    elif row == '25k_to_35k':
        return 1.3
    elif row == '100k_to_150k':
        return 0.60
    else: 
        return 0
```

```python
def age_func(row): 
    if row == '65-69':
        return 2.57
    elif row == '60-64':
        return 2.46
    elif row == '35-39':
        return 2.07
    elif row == '90 +':
        return 1.44
    elif row == '30-34':
        return 1.09
    elif row == '85-89':
        return 0.49
    elif row == '40-44' or row == '80-84' or row == '70-74' :
        return 0.41
    elif row == '45-49': 
        return 0.36
    elif row == '50-54':
        return 0.35
    else: 
        return 0
```

```python
df['income_score'] = df.Income_level.apply(income_func)
df['age_score'] = df.Age_range.apply(age_func)
df['unemply_score'] = df.Unemply_rate_range.apply(unemply_func)
df['college_score'] = df.Education_college.apply(edu_college_func)
df['highsch_score'] = df.Education_highschool.apply(edu_highschool_func)
df['labor_score'] = df.Labor_force_part_range.apply(labor_func)
```

```python
df['total_score'] = abs(df.income_score + df.age_score + df.unemply_score +
                    df.college_score + df.highsch_score + df.labor_score)
```

# Creating score group

```python
def score_func(row): 
    if row >= 25:
        return 'A'
    elif row < 25 and row >= 20:
        return 'Aa'
    elif row < 20 and row >= 15:
        return 'Aaa'
    elif row < 15 and row >= 13:
        return 'B'
    elif row < 13 and row >= 12:
        return 'Bb'
    elif row < 12 and row >= 10:
        return 'Bbb'
    elif row < 10 and row >= 8: 
        return 'C'
    elif row < 8 and row >= 6:
        return 'Cc'
    elif row < 6 and row >= 5:
        return 'D'
    elif row < 5 and row >= 4:
        return 'Dd'
    else: 
        return 'F'
```

# Group by `Grade`

```python
df['Grade'] = df.total_score.apply(score_func)
```

```python
paid = df[(df.Closure_Type == "PIF") | (df.Closure_Type == "SETTLE")]
unpaid = df[(df.Closure_Type != "PIF") & (df.Closure_Type != "SETTLE")]
```

```python
a = paid.groupby(['Grade'])['Total_Paid'].size().reset_index()
a.rename(columns={'Total_Paid':'Total_Paid_Count'}, inplace=True)
```

```python
b = unpaid.groupby(['Grade'])['Total_Paid'].size().reset_index()
b.rename(columns={'Total_Paid' : 'Unpaid_Count'}, inplace=True)
```

```python
ab = pd.merge(b, a, on=['Grade'])
ab['Total_Accounts'] = ab.Unpaid_Count + ab.Total_Paid_Count
ab['Percent_Paid'] = round((ab.Total_Paid_Count / ab.Total_Accounts)*100, 2)
ab['%_of_Paid'] = round((ab.Total_Paid_Count / ab.Total_Paid_Count.sum())*100, 2)
```

```python
c = df.groupby(by=['Grade'])['Invoice_bal'].mean().reset_index()
c.rename(columns={'Invoice_bal' : 'Avg_Inv_bal'}, inplace=True)
d = paid.groupby(by=['Grade'])['Invoice_bal'].mean().reset_index()
d.rename(columns={'Invoice_bal' : 'Paid_Avg_Inv_bal'}, inplace=True)
e = paid.groupby(by=['Grade'])['Total_Paid'].mean().reset_index()
e.rename(columns={'Total_Paid' : 'Avg_Collected'}, inplace=True)
f = paid.groupby(by=['Grade'])['Total_Paid'].sum().reset_index()
f.rename(columns={'Total_Paid' : 'Total_Collected'}, inplace=True)
g = df.groupby(by=['Grade'])['Invoice_bal'].sum().reset_index()
g.rename(columns={'Total_Paid' : 'Total_Inv_bal'}, inplace=True)

#merging datasets
cd = pd.merge(c,d, on='Grade')
cde = pd.merge( cd,e , on='Grade')
cdef = pd.merge(cde, f, on='Grade')
cdefg = pd.merge(cdef, g, on='Grade')
cdefg['%_Collected'] = round((cdefg.Total_Collected / cdefg.Invoice_bal) * 100, 2)

grade_groupby = pd.merge(ab, cdefg, on='Grade')
grade_groupby['%_of_Total_Accounts'] = round((grade_groupby.Total_Accounts / grade_groupby.Total_Accounts.sum()) * 100, 2)
```

# Group by `Grade` & `Age_range` & `Income_level`

```python
o = unpaid.groupby(['Grade','Age_range','Income_level'])['Total_Paid'].size().reset_index()
o.rename(columns={'Total_Paid' : 'Unpaid_Count'}, inplace=True)

p = paid.groupby(['Grade','Age_range','Income_level'])['Total_Paid'].size().reset_index()
p.rename(columns={'Total_Paid': 'Total_Paid_Count'}, inplace=True)

op = pd.merge(o,p, on=['Grade','Age_range','Income_level'])
```

```python
op['Total_Accounts'] = op.Unpaid_Count + op.Total_Paid_Count
op['Percent_Paid'] = round((op.Total_Paid_Count / op.Total_Accounts)*100, 2)
op['%_of_Paid'] = round((op.Total_Paid_Count / op.Total_Paid_Count.sum())*100, 2)

q = df.groupby(by=['Grade','Age_range','Income_level'])['Invoice_bal'].mean().reset_index()
q.rename(columns={'Invoice_bal' : 'Avg_Inv_bal'}, inplace=True)
r = paid.groupby(by=['Grade','Age_range','Income_level'])['Invoice_bal'].mean().reset_index()
r.rename(columns={'Invoice_bal' : 'Paid_Avg_Inv_bal'}, inplace=True)
s = paid.groupby(by=['Grade','Age_range','Income_level'])['Total_Paid'].mean().reset_index()
s.rename(columns={'Total_Paid' : 'Avg_Collected'}, inplace=True)
t = paid.groupby(by=['Grade','Age_range','Income_level'])['Total_Paid'].sum().reset_index()
t.rename(columns={'Total_Paid' : 'Total_Collected'}, inplace=True)
u = df.groupby(by=['Grade','Age_range','Income_level'])['Invoice_bal'].sum().reset_index()
u.rename(columns={'Total_Paid' : 'Total_Inv_bal'}, inplace=True)

#merging 
qr = pd.merge(q,r, on=['Grade','Age_range','Income_level'])
qrs = pd.merge(qr, s, on=['Grade','Age_range','Income_level'])
qrst = pd.merge(qrs, t, on=['Grade','Age_range','Income_level'])
qrstu = pd.merge(qrst, u, on=['Grade','Age_range','Income_level'])
```

```python
grade_age_income_groupby = pd.merge(op, qrstu, on=['Grade','Age_range','Income_level'])
grade_age_income_groupby['%_of_Total_Accounts'] = round((grade_age_income_groupby.Total_Accounts / grade_age_income_groupby.Total_Accounts.sum()) * 100, 2)
```
